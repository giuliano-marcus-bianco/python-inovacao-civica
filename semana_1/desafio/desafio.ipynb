# Explorando dados

Neste desafio colocaremos em prática alguns conceitos apresentado nas aulas. Para isso, usaremos o conjunto de dados [Candidatos - 2022](https://dadosabertos.tse.jus.br/dataset/candidatos-2022) disponível no Portal de Dados Abertos do Tribunal Superior Eleitoral.

Os dados também podem ser encontrados na pasta "Dados" deste repositório.

Siga as orientações abaixo para completar o desafio:

# Importe os pacotes necessários

```python
import pandas as pd
```

# Crie uma lista com o título dos arquivos usando a biblioteca Gloob

```python
import glob

arquivos_estados = glob.glob('base_de_dados/*.csv')
arquivos_estados
```

# Crie um laço (loop) para ler todos os arquivos da lista (usando a biblioteca Pandas) e armazenar os dataframes em uma lista

```python
lista_estados = []

for arquivo_estado in arquivos_estados:
    lista_estados.append(pd.read_csv(arquivo_estado, sep=';', encoding='latin_1'))

lista_estados[-1]
```

# Concatene todos os dataframes da lista

```python
df_estados = pd.concat(lista_estados, axis=0, ignore_index=True)
df_estados
```

# Apresente 5 linhas aleatórias do dataframe

```python
df_estados.sample(5)
```

# Quantas candidaturas estão em situação de reeleição?

```python
reeleicao = len(df_estados[df_estados.ST_REELEICAO == 'S'])
print(f'O número de candidados em situação de reeleição é igual a {reeleicao}.')
```

# Liste todas a ocupações declaradas pelas candidaturas

```python
ocupacoes = df_estados['DS_OCUPACAO'].unique()
print(f'A lista de ocupações declarada pelas candidaturas é a seguinte: {ocupacoes}.')
```

# Quantos anos tem a pessoa candidata mais velha?

```python
idade_max = df_estados.NR_IDADE_DATA_POSSE.max()
print(f'A pessoa candidata mais velha na data da posse teria {int(idade_max)} anos')
# df_estados.sort_values('NR_IDADE_DATA_POSSE', ascending=False).NM_CANDIDATO
```

# Quantos anos tem a pessoa candidata mais jovem?

```python
idade_min = df_estados.NR_IDADE_DATA_POSSE.min()
print(f'A pessoa candidata mais jovem na data da posse teria {int(idade_min)} anos')
```

# Qual é o nome da pessoa candidata mais velha?

```python
df_pessoa_idade_max = df_estados[df_estados['NR_IDADE_DATA_POSSE'] == idade_max]
nome_pessoa_idade_max = list(df_pessoa_idade_max['NM_CANDIDATO'])[0]

print(f'A pessoa candidata mais velha na data da posse se chama {nome_pessoa_idade_max}')
```

```python
candidata_mais_velha = df_estados[df_estados['NR_IDADE_DATA_POSSE'] == idade_max]['NM_CANDIDATO']
candidata_mais_velha
```

# Quantas candidaturas existem para "Deputado federal" seu estado?

```python
df_dep_federal = df_estados[df_estados['DS_CARGO'] == 'DEPUTADO FEDERAL']
df_dep_federal_sp = df_dep_federal[df_dep_federal['SG_UE'] == "SP"]
quantidade_dep_federal_sp = len(df_dep_federal_sp)
print(f'A quantidade de candidatura para Deputado Federal no estado de São Paulo é igual a {quantidade_dep_federal_sp}.')
```

# Qual é a média do valor máximo de despesas de campanha?

```python
media_valor_max_despesa = df_estados['VR_DESPESA_MAX_CAMPANHA'].mean()
print(f'A média do valor máximo de despesas de campanha é igual R$ {media_valor_max_despesa:.2f}')
```
