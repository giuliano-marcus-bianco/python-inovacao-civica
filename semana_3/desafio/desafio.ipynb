### 2) Leia o Arquivo csv 

```python
import pandas as pd

pd.set_option("display.max_columns", None)

DTYPE = {
    "txNomeParlamentar": str,
    "cpf": str,
    "ideCadastro": str,
    "nuCarteiraParlamentar": str,
    "nuLegislatura": str,
    "sgUF": str,
    "sgPartido": str,
    "codLegislatura": str,
    "numSubCota": str,
    "txtDescricao": str,
    "numEspecificacaoSubCota": str,
    "txtDescricaoEspecificacao": str,
    "txtFornecedor": str,
    "txtCNPJCPF": str,
    "txtNumero": str,
    "indTipoDocumento": str,
    "datEmissao": str,
    "vlrDocumento": float,
    "vlrGlosa": str,
    "vlrLiquido": float,
    "numMes": str,
    "numAno": str,
    "numParcela": str,
    "txtPassageiro": str,
    "txtTrecho": str,
    "numLote": str,
    "numRessarcimento": str,
    "nuDeputadoId": str,
    "ideDocumento": str,
}

df2018 = pd.read_csv(f"http://www.camara.leg.br/cotas/Ano-2018.csv.zip", sep=";", compression="zip", dtype=DTYPE)
```

#### [Desafio Extra] Existem registros de outros anos no arquivo de 2018? Se sim, quais anos e quantos registros há nessas datas?

### 6) Quais os meses que mais tem solicitação de reembolso?

```python
df2018.datEmissao.dt.month_name().rename("Meses").value_counts(dropna=False).to_frame("Contagem")
```

#### [Desafio Extra] Como seria para agrupar os dados no formato `mes/ano` , ao invés de apenas mês?

```python
df2018["datEmissao"].groupby(by=[df2018["datEmissao"].dt.year.rename("Anos"), df2018["datEmissao"].dt.month.rename("Meses")], dropna=False).count().to_frame("Contagem")
```

### 7) Limpe a coluna `txtCNPJCPF`: ela deve conter somente dígitos numéricos

```python
# vamos trocar qualquer caracter que não seja um dígito número por '', essa operação remove os dígitos

# df_reimbursements["txtCNPJCPF"] = df_reimbursements...
df2018.txtCNPJCPF = df2018.txtCNPJCPF.str.replace(r"\D", "", regex=True)
df2018.txtCNPJCPF.sample(30)
```

### 8) Quais são os valores únicos do campo `indTipoDocumento`? Substitua o conteúdo pela respectiva legenda

*Dica*: 0 (Zero), para Nota Fiscal; 1 (um), para Recibo; e 2, para Despesa no Exterior.

```python
df2018.indTipoDocumento.unique()
```

```python
converters = {"0": "nota_fiscal", "1": "recibo", "2": "despesa_exterior"}
df2018.indTipoDocumento.replace(converters, inplace=True)
df2018.indTipoDocumento.unique()
```

### [Desafio Extra] 9) Agrupe o valor total de despesas por fornecedor, mês e ano. Você encontrou algum padrão?

```python
df2018["vlrLiquido"].groupby(by=[df2018.txtFornecedor.rename("Fornecedor"), df2018.datEmissao.dt.month.rename("Mês"),  df2018.datEmissao.dt.year.rename("Ano")]).sum().sort_values(ascending=False).to_frame("Valor Líquido").head(15)
```

### [Desafio Extra] 10) Quais os 10 parlamentares que mais solicitaram reembolso?

```python
df2018.txNomeParlamentar.groupby(by=df2018.txNomeParlamentar.rename("Nomes")).count().to_frame("Contagem").sort_values(by="Contagem", ascending=False).head(10)
```

#### [Desafio Extra] Como ficaria a resposta, caso queiramos saber quais os seus estados e partidos?

```python
df2018.txNomeParlamentar.groupby(
    by=[df2018.txNomeParlamentar.rename("Nomes"),
    df2018.sgUF.rename("Estado"),
    df2018.sgPartido.rename("Partido")]
    ).count().to_frame("Contagem").sort_values(by="Contagem", ascending=False).head(10)
```

### [Desafio Extra] 11) Quais os partidos e os estados que mais solicitaram reembolso?

```python
df2018.sgUF.groupby(
    by=[df2018.sgPartido.rename("Partidos"),
        df2018.sgUF.rename("Estados"),]
).count().to_frame("Contagem").sort_values(by="Contagem", ascending=False).head(30)
```

### [Desafio Extra] 12) Recupere os reembolsos de 2019 e acrescente aos dados de 2018

```python
# Baixando os dados de 2019
df2019 = pd.read_csv(f"http://www.camara.leg.br/cotas/Ano-2019.csv.zip", sep=";", compression="zip", dtype=DTYPE)

# Limpando os dados de 2019
df2019.datEmissao = pd.to_datetime(df2019.datEmissao, format="%Y-%m-%dT%H:%M:%S") # ajustando a coluna "datEmissao" para datetime

df2019.txtCNPJCPF = df2019.txtCNPJCPF.str.replace(r"\D", "", regex=True) # limpando caracteres da coluna "txtCNPJCPF" que não sejam dígitos

df2019.indTipoDocumento.replace(converters, inplace=True) # convertendo os dígitos da coluna "indTipoDocumento" para seus nomes por extenso
```

```python
# concatenando os dataframes de 2018 e 2019
df = pd.concat(objs=[df2018, df2019], ignore_index=True, keys=["2018", "2019"])
```
